```python
lst=[]
with open('Input\data_source_2\sample_data.3.dat','r') as dat_file:
    for line in dat_file:
        lst.append(line)
```

```python
lst
```

```python
lst2 = []
for line in lst:
    lst2.append(line.split(','))
    
```

```python
lst2
```

```python
for line in lst2:
    line[-1] = line[-1].replace('\n','')    
```

```python
lst2
```

```python
df3=pd.DataFrame(lst2[1:], columns=lst2[0])
```

```python
df3
```

```python
df3['source']='data_source_2'
```

```python
df3
```

```python
df_final=pd.concat([df1,df2,df3], ignore_index = True)
```
